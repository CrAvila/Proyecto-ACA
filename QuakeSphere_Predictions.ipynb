```python
import requests
import pandas as pd
import json
from datetime import datetime

# Function to fetch earthquake data for a given year
def fetch_earthquake_data(year):
    base_url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
    params = {
        'format': 'geojson',
        'starttime': f'{year}-01-01',
        'endtime': f'{year}-12-31',
        'minmagnitude': '2.5',
        'orderby': 'time',
        'limit': 20000,
        'offset': 1
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        return data['features']
    else:
        print(f"Error fetching data for year {year}: {response.status_code}")
        return []

# Function to parse earthquake data into a DataFrame
def parse_earthquake_data(features):
    records = []
    for feature in features:
        properties = feature['properties']
        geometry = feature['geometry']
        record = {
            'time': properties['time'],
            'latitude': geometry['coordinates'][1],
            'longitude': geometry['coordinates'][0],
            'depth': geometry['coordinates'][2],
            'mag': properties['mag'],
            'place': properties['place'],
            'type': properties['type']
        }
        records.append(record)
    return pd.DataFrame(records)

# Fetch and combine earthquake data for multiple years
start_year = 2000
end_year = datetime.now().year
all_features = []

for year in range(start_year, end_year + 1):
    features = fetch_earthquake_data(year)
    all_features.extend(features)
    print(f"Year {year}: Fetched {len(features)} earthquakes")

# Parse all data into a DataFrame
df = parse_earthquake_data(all_features)

# Save DataFrame to CSV
df.to_csv('all_earthquakes.csv', index=False)
print(f"Total earthquakes fetched: {len(df)}")
```
